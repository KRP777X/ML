## Reshaping

Arrays can be reshaped. We will start with an example. Let us start with a matrix $\mathbf{M}$:

$$
\mathbf{M} = \begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6
\end{bmatrix}
$$

We can now reshape it into a vector:

$$
\mathbf{x} = \begin{bmatrix}
1 & 2 & 3 & 4 & 5 & 6
\end{bmatrix}
$$

In this course, we will mainly stick to matrix-vector addition or multiplication. This is the recipe:

- Row-wise
  - `M + b`
- Column-wise: 
  - Make `b` a column vector or a `(m, 1)` matrix using `np.expand_dims`
  - `M + b`

## Stacking arrays

Sometimes, we would want to stack arrays. Consider the two matrices:

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2\\
3 & 4
\end{bmatrix},
\mathbf{B} =
\begin{bmatrix}
5 & 6\\
7 & 8
\end{bmatrix}
$$

There are two ways to stack these two matrices:

### Row-wise

We could stack the two matrices along the rows, $\mathbf{A}$ on top of $\mathbf{B}$:

$$
\mathbf{C} =
\begin{bmatrix}
1 & 2\\
3 & 4\\
5 & 6\\
7 & 8
\end{bmatrix}
$$

This would be done as follows:

by the concactenate function:-

for row wise- `C = np.concatenate((A, B), axis = 0)`

fo column wise- `C = np.concatenate((A, B), axis = 1)`

Define a function ConfusionMatrix(y,y_hat) for binary classification and return a matrix in the following format: [[TN,FP],[FN,TP]]

Inputs: y: 1 x n, y_hat: 1 x n

Output: 2D numpy array.

```python
import numpy as np
def ConfusionMatrix(y_hat,y):
  TP=np.where((y==1)&(y_hat==1),1,0).sum()
  TN=np.where((y==0)&(y_hat==0),1,0).sum()
  FP=np.where((y==0)&(y_hat==1),1,0).sum()
  FN=np.where((y==1)&(y_hat==0),1,0).sum()

  return np.array([TN,FP],[FN,TP])
```

the sample in X and return the loss for individual samples in a vector.

```python
def percep_loss(x,w,y):
  z=x@w
  y_predict=np.where(z>=0,-1,1)
  samplewise_loss=np.maximum(-1*y_predict*y,np.zeros(y.shape[0]))
  return samplewise_loss
```

Tensorflow

```python
import tensorflow as tf
```

```python
x=tf.Variable(5.0)
```

```python
with tf.GradientTape() as t:
  t.watch(x)   #for tracking forward pass
  y=x**2
```

```python
t.gradient(y,x) #for using autodiff
```
